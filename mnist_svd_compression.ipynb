```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import os
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model, clone_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input, BatchNormalization, Activation, Add
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import Callback
from tensorflow.keras.mixed_precision import set_global_policy

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Enable mixed precision training
set_global_policy('mixed_float16')

def load_and_preprocess_mnist_data(train_path, test_path):
    """
    Load and preprocess the MNIST dataset following the implementation steps

    Args:
        train_path (str): Path to training data CSV
        test_path (str): Path to test data CSV

    Returns:
        Preprocessed training, validation, and test data
    """
    # Analysis Phase: Load and preprocess the MNIST dataset
    train = pd.read_csv(train_path)
    test = pd.read_csv(test_path)

    # Extract images and labels
    X = train.drop("label", axis=1).values.astype(np.float32)
    y = train["label"].values
    X_test = test.values.astype(np.float32)

    # Normalize pixel values to [0, 1]
    X = X / 255.0
    X_test = X_test / 255.0

    # Convert labels to one-hot encoding
    num_classes = len(np.unique(y))
    y_cat = to_categorical(y, num_classes)

    # Reshape images to 28x28x1
    X = X.reshape(-1, 28, 28, 1)
    X_test = X_test.reshape(-1, 28, 28, 1)

    # Split training data into train and validation sets
    X_train, X_val, y_train, y_val = train_test_split(X, y_cat, test_size=0.2, random_state=42)

    return X_train, y_train, X_val, y_val, X_test

def create_mini_resnet(input_shape=(28, 28, 1), num_classes=10):
    """
    Create a mini-ResNet model to establish baseline performance

    Args:
        input_shape (tuple): Input image shape
        num_classes (int): Number of output classes

    Returns:
        Compiled Keras model
    """
    inputs = Input(shape=input_shape)

    # Initial convolutional layer
    x = Conv2D(32, 3, padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # First stack of residual blocks
    x = residual_block(x, 32)
    x = residual_block(x, 32)

    # Downsampling with stride 2
    x = residual_block(x, 64, stride=2, conv_shortcut=True)
    x = residual_block(x, 64)

    # Global average pooling and final dense layer
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu', use_bias=False)(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    # Create model
    model = Model(inputs=inputs, outputs=outputs)

    # Use standard Adam optimizer with tuned parameters
    optimizer = Adam(
        learning_rate=1e-3,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7
    )

    # Compile with mixed precision
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        jit_compile=True
    )

    return model

def residual_block(x, filters, kernel_size=3, stride=1, conv_shortcut=False):
    """
    Create a residual block with two convolutional layers and a skip connection

    Args:
        x: Input tensor
        filters: Number of filters in the convolutional layers
        kernel_size: Size of the kernel in the convolutional layers
        stride: Stride for the first convolutional layer
        conv_shortcut: Whether to use a convolutional shortcut or an identity shortcut

    Returns:
        Output tensor for the residual block
    """
    # Store input
    shortcut = x

    # First convolutional layer
    x = Conv2D(filters, kernel_size, strides=stride, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Second convolutional layer
    x = Conv2D(filters, kernel_size, padding='same')(x)
    x = BatchNormalization()(x)

    # Shortcut connection
    if conv_shortcut:
        shortcut = Conv2D(filters, 1, strides=stride, padding='same')(shortcut)
        shortcut = BatchNormalization()(shortcut)

    # Add shortcut to output
    x = Add()([x, shortcut])
    x = Activation('relu')(x)

    return x

def get_model_size(model):
    """
    Measure model size in MB

    Args:
        model: Keras model

    Returns:
        Model size in MB
    """
    # Save model to a temporary file
    temp_file = 'temp_model.h5'
    model.save(temp_file)
    # Get size of file in bytes
    size_bytes = os.path.getsize(temp_file)
    # Convert to MB
    size_mb = size_bytes / (1024 * 1024)
    # Remove temporary file
    os.remove(temp_file)
    return size_mb

def measure_inference_time(model, test_data, num_samples=100):
    """
    Measure inference time for the model

    Args:
        model: Keras model
        test_data: Test dataset
        num_samples: Number of samples to use for measurement

    Returns:
        Average inference time per sample in milliseconds
    """
    # Select a subset of test data for measurement
    subset = test_data[:min(num_samples, len(test_data))]

    # Warm-up run
    _ = model.predict(subset[:10])

    # Measure inference time
    start_time = time.time()
    _ = model.predict(subset)
    end_time = time.time()

    # Calculate average inference time per sample in milliseconds
    avg_time_ms = (end_time - start_time) * 1000 / len(subset)
    return avg_time_ms

class MetricsTracker(tf.keras.callbacks.Callback):
    """
    Custom callback to track model size and inference time
    """
    def __init__(self, validation_data):
        super().__init__()
        self.validation_data = validation_data
        self.model_sizes = []
        self.inference_times = []

    def on_epoch_end(self, epoch, logs=None):
        # Calculate model size
        model_size = get_model_size(self.model)
        self.model_sizes.append(model_size)

        # Measure inference time
        inference_time = measure_inference_time(self.model, self.validation_data[0])
        self.inference_times.append(inference_time)

        print(f"Epoch {epoch+1} - Model size: {model_size:.2f} MB, Inference time: {inference_time:.2f} ms per sample")

def find_conv_layers(model):
    """
    Find convolutional layers in the model

    Args:
        model: Keras model

    Returns:
        List of tuples with layer index, name, layer, and parameter count
    """
    conv_layers = []
    for i, layer in enumerate(model.layers):
        if isinstance(layer, Conv2D):
            # Calculate parameters
            params = np.prod(layer.kernel.shape)
            if layer.use_bias:
                params += layer.filters
            conv_layers.append((i, layer.name, layer, params))
    return conv_layers

def visualize_model_architecture(model):
    """
    Visualize the model architecture and highlight potential SVD compression candidates

    Args:
        model: Keras model

    Returns:
        Sorted list of convolutional layers
    """
    conv_layers = find_conv_layers(model)

    # Calculate parameters per layer
    layer_params = []
    for i, layer in enumerate(model.layers):
        if hasattr(layer, 'trainable_weights') and layer.trainable_weights:
            params = np.sum([np.prod(w.shape) for w in layer.trainable_weights])
            layer_params.append((i, layer.name, params, type(layer).__name__))

    # Sort by number of parameters
    layer_params.sort(key=lambda x: x[2], reverse=True)

    # Plot parameters per layer
    plt.figure(figsize=(15, 10))

    indices = [x[0] for x in layer_params]
    names = [f"{x[0]}: {x[1]}" for x in layer_params]
    params = [x[2] for x in layer_params]
    types = [x[3] for x in layer_params]

    # Create bar plot
    bars = plt.bar(names, params)

    # Highlight conv layers
    conv_indices = [x[0] for x in conv_layers]
    for i, bar in enumerate(bars):
        if indices[i] in conv_indices:
            bar.set_color('orange')
        else:
            bar.set_color('blue')

    plt.title('Parameters Per Layer - Orange Bars = Conv2D Layers (SVD Candidates)', fontsize=14)
    plt.xlabel('Layer (index: name)')
    plt.ylabel('Number of Parameters')
    plt.xticks(rotation=90)
    plt.grid(axis='y')

    # Add layer type annotations
    for i, bar in enumerate(bars):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'{types[i]}', ha='center', va='bottom', rotation=90)

    plt.tight_layout()
    plt.show()

    # Print summary of convolutional layers
    print("\nConvolutional layers sorted by parameter count:")
    conv_layers_sorted = sorted(conv_layers, key=lambda x: x[3], reverse=True)
    for i, name, layer, params in conv_layers_sorted:
        print(f"Layer {i}: {name} - Shape: {layer.kernel.shape} - Parameters: {params}")

    return conv_layers_sorted

def compress_layer_with_svd(model, layer_index, energy_retention_threshold=0.9):
    """
    Compress a layer using Singular Value Decomposition (SVD)

    Args:
        model: Original Keras model
        layer_index: Index of the layer to compress
        energy_retention_threshold: Percentage of energy to retain

    Returns:
        Compressed model and compression information
    """
    # Clone the original model
    compressed_model = clone_model(model)
    compressed_model.set_weights(model.get_weights())

    # Identify the layer to compress
    layer = model.layers[layer_index]

    # Only compress Conv2D and Dense layers
    if not isinstance(layer, (Conv2D, Dense)):
        return model, None

    # Get the weights of the target layer
    weights = layer.get_weights()
    W = weights[0]  # Weight matrix
    b = weights[1] if len(weights) > 1 else None  # Bias (optional)

    # Reshape weights for SVD decomposition
    original_shape = W.shape
    original_params = np.prod(original_shape)

    if isinstance(layer, Conv2D):
        # For Conv2D, reshape to 2D matrix with (kernel_height * kernel_width * input_channels, output_channels)
        W_reshaped = W.reshape((-1, original_shape[-1]))
    else:
        # For Dense, already 2D
        W_reshaped = W

    # Perform SVD
    U, S, V = np.linalg.svd(W_reshaped, full_matrices=False)

    # Calculate cumulative energy and determine rank
    energy = S**2
    total_energy = np.sum(energy)
    cumulative_energy = np.cumsum(energy) / total_energy

    # Find rank that retains specified energy
    k = np.argmax(cumulative_energy >= energy_retention_threshold) + 1

    # Print SVD analysis
    print(f"\nLayer {layer_index} ({layer.name}) SVD Analysis:")
    print(f"Original shape: {original_shape}, Total parameters: {original_params:,}")
    print(f"Singular values range: min={S[-1]:.4f}, max={S[0]:.4f}, ratio={S[-1]/S[0]:.4f}")
    print(f"Selected rank: {k} (retaining {cumulative_energy[k-1]:.2%} energy)")

    # Truncate matrices
    U_k = U[:, :k]
    S_k = np.diag(S[:k])
    V_k = V[:k, :]

    # Reconstruct weight matrix
    W_compressed = U_k @ S_k @ V_k

    # Reshape back to original shape
    if isinstance(layer, Conv2D):
        # Reshape back to conv2d kernel shape
        W_compressed = W_compressed.reshape(original_shape)
    else:
        # For dense layers, already in correct shape
        W_compressed = W_compressed

    # Prepare new weights
    if b is not None:
        new_weights = [W_compressed, b]
    else:
        new_weights = [W_compressed]

    # Find the correct weights index in the model
    all_weights = compressed_model.get_weights()

    # Update weights for the specific layer
    layer_weights_indices = [i for i, w in enumerate(all_weights) if np.array_equal(w, weights[0])]

    if layer_weights_indices:
        weight_index = layer_weights_indices[0]
        # Replace the weights while maintaining the overall structure
        all_weights[weight_index:weight_index+len(new_weights)] = new_weights
    else:
        # Fallback method if exact match not found
        for i in range(len(all_weights)):
            try:
                all_weights[i] = new_weights[0] if len(new_weights) == 1 else new_weights
                break
            except:
                continue

    # Set updated weights
    compressed_model.set_weights(all_weights)

    # Recompile the model
    compressed_model.compile(
        optimizer=Adam(),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Calculate compression parameters
    if isinstance(layer, Conv2D):
        compressed_params = (original_shape[0] * original_shape[1]) * k + k * original_shape[-1]
    else:
        compressed_params = original_shape[0] * k + k * original_shape[1]

    # Compression information
    compression_info = {
        'layer_index': layer_index,
        'layer_name': layer.name,
        'original_params': original_params,
        'compressed_params': compressed_params,
        'actual_compression': compressed_params / original_params,
        'singular_values_kept': k,
        'total_singular_values': len(S),
        'singular_values': S,
        'energy_retained': cumulative_energy[k-1]
    }

    			# === Compute Computational Expense ===
		    # === Compute Computational Expense ===
    if isinstance(layer, Conv2D):
        # Extract kernel and channels
        kh, kw, Cin, Cout = original_shape
        input_shape = layer.input.shape[1:]
        output_shape = layer.output.shape[1:]


        H_out, W_out = output_shape[0], output_shape[1]
        original_flops = H_out * W_out * Cin * Cout * kh * kw
        compressed_flops = H_out * W_out * k * (Cin + Cout)
    else:
        # Dense layer
        Cin, Cout = original_shape
        original_flops = 2 * Cin * Cout
        compressed_flops = 2 * (Cin + Cout) * k

    # Computational intensity (FLOPs per parameter)
    original_intensity = original_flops / original_params
    compressed_intensity = compressed_flops / compressed_params
    efficiency_ratio = original_intensity / compressed_intensity

    # Add to compression info
    compression_info.update({
        'original_flops': original_flops,
        'compressed_flops': compressed_flops,
        'flops_reduction': 1 - compressed_flops / original_flops,
        'original_intensity': original_intensity,
        'compressed_intensity': compressed_intensity,
        'efficiency_ratio': efficiency_ratio
    })

    # Print computational expense summary
    print(f"\n=== Computational Expense Analysis for Layer {layer.name} ===")
    print(f"  Original FLOPs: {original_flops:,}")
    print(f"  Compressed FLOPs: {compressed_flops:,}")
    print(f"  FLOPs Reduction: {compression_info['flops_reduction']*100:.2f}%")
    print(f"  Original Computational Intensity: {original_intensity:.2f} FLOPs/param")
    print(f"  Compressed Computational Intensity: {compressed_intensity:.2f} FLOPs/param")
    print(f"  Computational Efficiency Ratio: {efficiency_ratio:.2f}x")


    print(f"Original parameters: {original_params:,}")
    print(f"Compressed parameters: {compressed_params:,}")
    print(f"Actual compression ratio: {compression_info['actual_compression']:.2%}")
    print(f"Energy retained: {compression_info['energy_retained']:.2%}")

    return compressed_model, compression_info

def visualize_compression_comparison(original_history, compressed_history,
                                     original_metrics, compressed_metrics,
                                     compression_infos):
    """
    Creates detailed visualizations comparing original and compressed models

    Args:
        original_history: Training history of the original model
        compressed_history: Training history of the compressed model
        original_metrics: MetricsTracker instance for the original model
        compressed_metrics: MetricsTracker instance for the compressed model
        compression_infos: List of compression info dictionaries
    """
    # Create figure with subplots
    fig, axes = plt.subplots(3, 2, figsize=(16, 15))

    # Plot accuracy
    axes[0, 0].plot(original_history.history['accuracy'], 'b-', label='Original Train')
    axes[0, 0].plot(original_history.history['val_accuracy'], 'b--', label='Original Val')
    if compressed_history:
        axes[0, 0].plot(compressed_history.history['accuracy'], 'r-', label='Compressed Train')
        axes[0, 0].plot(compressed_history.history['val_accuracy'], 'r--', label='Compressed Val')
    axes[0, 0].set_title('Accuracy Comparison')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Plot loss
    axes[0, 1].plot(original_history.history['loss'], 'b-', label='Original Train')
    axes[0, 1].plot(original_history.history['val_loss'], 'b--', label='Original Val')
    if compressed_history:
        axes[0, 1].plot(compressed_history.history['loss'], 'r-', label='Compressed Train')
        axes[0, 1].plot(compressed_history.history['val_loss'], 'r--', label='Compressed Val')
    axes[0, 1].set_title('Loss Comparison')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Loss')
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # Plot model size
    epochs_orig = range(1, len(original_metrics.model_sizes) + 1)
    axes[1, 0].plot(epochs_orig, original_metrics.model_sizes, 'bo-', label='Original')
    if compressed_metrics:
        epochs_comp = range(1, len(compressed_metrics.model_sizes) + 1)
        axes[1, 0].plot(epochs_comp, compressed_metrics.model_sizes, 'ro-', label='Compressed')
    axes[1, 0].set_title('Model Size Comparison')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Size (MB)')
    axes[1, 0].legend()
    axes[1, 0].grid(True)

    # Plot inference time
    axes[1, 1].plot(epochs_orig, original_metrics.inference_times, 'bo-', label='Original')
    if compressed_metrics:
        axes[1, 1].plot(epochs_comp, compressed_metrics.inference_times, 'ro-', label='Compressed')
    axes[1, 1].set_title('Inference Time Comparison')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Time per sample (ms)')
    axes[1, 1].legend()
    axes[1, 1].grid(True)

    # Plot singular value distribution for compressed layers
    if compression_infos:
        ax = axes[2, 0]
        colors = ['r', 'g', 'b', 'c', 'm']
        for i, info in enumerate(compression_infos):
            color = colors[i % len(colors)]
            S = info['singular_values']
            ax.semilogy(range(len(S)), S, color=color,
                      label=f"Layer {info['layer_name']}")
            ax.axvline(x=info['singular_values_kept'], color=color, linestyle='--')
        ax.set_title('Singular Value Distribution (Log Scale)')
        ax.set_xlabel('Index')
        ax.set_ylabel('Singular Value')
        ax.legend()
        ax.grid(True)

    # Plot parameter reduction
    if compression_infos:
        ax = axes[2, 1]
        layers = [info['layer_name'] for info in compression_infos]
        original = [info['original_params'] for info in compression_infos]
        compressed = [info['compressed_params'] for info in compression_infos]

        x = np.arange(len(layers))
        width = 0.35

        ax.bar(x - width/2, original, width, label='Original', color='blue')
        ax.bar(x + width/2, compressed, width, label='Compressed', color='orange')

        ax.set_title('Parameter Reduction by Layer')
        ax.set_ylabel('Number of Parameters')
        ax.set_xticks(x)
        ax.set_xticklabels(layers, rotation=45, ha='right')

        # Add compression percentage labels
        for i, info in enumerate(compression_infos):
            ratio = info['compressed_params'] / info['original_params'] * 100
            ax.text(i, original[i] / 2, f"{ratio:.1f}%", ha='center', color='white', fontweight='bold')

        ax.legend()
        ax.grid(True, axis='y')

    plt.tight_layout()
    plt.show()

def main():
    # Analysis Phase
    print("Loading and preprocessing MNIST dataset...")
    train_path = 'train.csv'
    test_path = 'test.csv'
    X_train, y_train, X_val, y_val, X_test = load_and_preprocess_mnist_data(train_path, test_path)

    # Create and train baseline mini-ResNet model
    print("\nCreating baseline mini-ResNet model...")
    baseline_model = create_mini_resnet()

    # Visualize model architecture
    print("\nVisualizing model architecture...")
    conv_layers = visualize_model_architecture(baseline_model)

    # Configure training
    BATCH_SIZE = 128
    NUM_EPOCHS = 5

    # Metrics Tracker for baseline model
    baseline_metrics_tracker = MetricsTracker(validation_data=(X_val, y_val))

    # Train baseline model
    print("\nTraining baseline model...")
    baseline_history = baseline_model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=NUM_EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[baseline_metrics_tracker],
        verbose=1
    )

    # Baseline performance
    baseline_loss, baseline_accuracy = baseline_model.evaluate(X_val, y_val)
    print(f"\nBaseline Model Performance:")
    print(f"Validation Accuracy: {baseline_accuracy:.4f}")

    # Compression Experiments
    compression_experiments = [
        {'threshold': 0.9, 'name': '90% Energy Retention'},
        {'threshold': 0.75, 'name': '75% Energy Retention'}
    ]

    # Store results of compression experiments
    experiment_results = []

    # Perform compression experiments
    for exp in compression_experiments:
        print(f"\n--- {exp['name']} Experiment ---")

        # Clone baseline model
        compressed_model = clone_model(baseline_model)
        compressed_model.set_weights(baseline_model.get_weights())

        # Layers to compress (select from top 3 conv layers)
        layers_to_compress = [
            conv_layers[0][0],  # Largest conv layer
            conv_layers[1][0],  # Second largest conv layer
            conv_layers[2][0]   # Third largest conv layer
        ]

        # Store compression information
        compression_infos = []

        # Compress selected layers
        for layer_idx in layers_to_compress:
            compressed_model, compression_info = compress_layer_with_svd(
                compressed_model,
                layer_idx,
                energy_retention_threshold=exp['threshold']
            )

            if compression_info:
                compression_infos.append(compression_info)

        # Metrics Tracker for compressed model
        compressed_metrics_tracker = MetricsTracker(validation_data=(X_val, y_val))

        # Fine-tune compressed model
        print("\nFine-tuning compressed model...")
        compressed_history = compressed_model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=NUM_EPOCHS,
            batch_size=BATCH_SIZE,
            callbacks=[compressed_metrics_tracker],
            verbose=1
        )

        # Evaluate compressed model
        comp_loss, comp_accuracy = compressed_model.evaluate(X_val, y_val)

        # Store experiment results
        experiment_results.append({
            'name': exp['name'],
            'baseline_accuracy': baseline_accuracy,
            'compressed_accuracy': comp_accuracy,
            'accuracy_drop': baseline_accuracy - comp_accuracy,
            'compression_infos': compression_infos,
            'baseline_history': baseline_history,
            'compressed_history': compressed_history,
            'baseline_metrics': baseline_metrics_tracker,
            'compressed_metrics': compressed_metrics_tracker
        })

        # Visualize compression results
        visualize_compression_comparison(
            baseline_history,
            compressed_history,
            baseline_metrics_tracker,
            compressed_metrics_tracker,
            compression_infos
        )

    # Print summary of experiments
    print("\n--- Compression Experiment Summary ---")
    for result in experiment_results:
        print(f"\n{result['name']}:")
        print(f"Baseline Accuracy: {result['baseline_accuracy']:.4f}")
        print(f"Compressed Accuracy: {result['compressed_accuracy']:.4f}")
        print(f"Accuracy Drop: {result['accuracy_drop']:.4f}")

        # Print layer compression details
        for info in result['compression_infos']:
            print(f"  {info['layer_name']}:")
            print(f"    Original Parameters: {info['original_params']:,}")
            print(f"    Compressed Parameters: {info['compressed_params']:,}")
            print(f"    Compression Ratio: {info['actual_compression']:.2%}")
            print(f"    Energy Retained: {info['energy_retained']:.2%}")

if __name__ == "__main__":
    main()
```
